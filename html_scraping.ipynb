```python
from bs4 import BeautifulSoup
import pandas
```

```python
opinions = pandas.read_csv("opinions_sample.csv", sep=";")
list(opinions)
```

```python
columns = ['case_cite','parties','docket','date','prelims','opinion_list']
extracted_data = pandas.DataFrame(columns=columns)
```

```python
for i in range(len(opinions.html)):
    case_cite, parties, docket, date, prelims, opinion_list = (list(),)*6
    try:
        soup = BeautifulSoup(opinions.html[i],"html.parser")
        try:
            case_cite = soup.findAll("p", {"class": "case_cite"})
        except:
            print("No Case Cite")
        try:
            parties = soup.findAll("p", {"class": "parties"})
        except:
            print("No parties")
        try:
            docket = soup.findAll("p", {"class": "docket"})
        except:
            print("No docket")
        try:
            date = soup.findAll("p", {"class": "date"})
        except:
            print("No date")
        try:
            prelims = soup.findAll("p", {"class": "prelims"})
        except:
            print("No prelims")
        try:
            opinion_list = soup.findAll("p", {"class": "opinions"})
        except:
            print("No opinions")
        extracted_data.loc[i] = [[obs.get_text() for obs in case_cite],[obs.get_text() for obs in parties],[obs.get_text() for obs in docket],[obs.get_text() for obs in date],[obs.get_text() for obs in prelims],[obs.get_text() for obs in opinion_list]]
    except:
        print("Nan HTML")
```
